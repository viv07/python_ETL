```python
df_emp=pd.read_csv(r'D:\python coding\pandas practice\emp.csv')
df_emp
```

```python
df_emp.dtypes
```

```python
#group by deptno and sum of sal
df_agg=df_emp.groupby(['DEPTNO'])['SAL'].mean()
df_agg
```

```python
df_comm=df_emp.fillna({'COMM':0})
df_comm
job_dict={'CLERK':'CLK','SALESMAN':'MKT','MANAGER':'MGR','ANALYST':'AST','PRESIDENT':'PREZ'}
df_new=df_comm.replace({'JOB':job_dict})
df_new
```

```python
#add a new column to df with mathematical calculation
df_new['comm_per']=(df_new['COMM']/df_new['SAL'])*100
df_new
```

```python
def emp_taxes(sal):
    tax=0
    if sal in df_new.iloc[:,5].to_list():
        if sal>500 and sal<=1400:
            tax=sal*.02
        if sal>1400 and sal<=2200:
            tax=sal*.03
        if sal>2200 and sal<=5000:
            tax=sal*.045
    return tax
```

```python
emp_taxes(5000)
```

```python
df_new['Tax_on_Sal']=df_new['SAL'].apply(emp_taxes)
```

```python
df_new
```

```python
df_dept=pd.read_csv(r'D:\python coding\pandas practice\dept.csv')
df_dept
```

```python
dname_dict={'ACCOUNTING':'ACCT','RESEARCH':'R&D','SALES':'M&S','OPERATIONS':'OPS'}
loc_dict={'NEW YORK':'NY','DALLAS':'DALI','CHICAGO':'CHI','BOSTON':'BOS'}
df_dname=df_dept.replace({'DNAME':dname_dict})
df_dname
df_loc=df_dname.replace({'LOC':loc_dict})
df_loc
```

```python
df_new.columns
#df_agg.columns
df_loc.columns
```

```python
#join dataframes
#df to be joined are df_loc, df_new and df_agg
j1=pd.merge(df_new[['EMPNO','ENAME', 'JOB','SAL', 'COMM', 'DEPTNO',
                    'comm_per', 'Tax_on_Sal']],df_agg,on='DEPTNO')
final_df=pd.merge(j1,df_loc[['DEPTNO','DNAME', 'LOC']],on='DEPTNO')
final_df
```

```python
#rename columns in final dataframe for better presentation
final_df.rename(columns={'SAL_x':'Salary','comm_per':'COMM_PER',
                'Tax_on_Sal':'Taxes Paid','SAL_y':'Avg Dept Salary'
                },inplace=True)
final_df
```
