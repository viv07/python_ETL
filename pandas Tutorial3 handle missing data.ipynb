```python
#conversion to date
df=pd.read_csv(r'D:\python coding\pandas practice\py-master\py-master\pandas\5_handling_missing_data_fillna_dropna_interpolate\weather_data.csv',
               parse_dates=['day'])

type(df.day[0])
#datatype of day is converted to timesatmp, prev it was string
```

```python
#cleansing null data
#playing around with the nulls in df
new_df=df.fillna(0)
new_df
#fillna replaces all the nulls to 0
#irrespective of the datatypes
```

```python
#filling nulls with column specific values
#using dictionary 
new_df=df.fillna({'temperature':0,
                 'windspeed':0,
                 'event':'event not recorded'})
new_df
```

```python
#now issue with above solution is that
#temp 0 in 2nd record look unrealistic
#so we have another way of approach to handle nulls
new_df=df.fillna(method='ffill')
new_df
#method= bfill,ffill,pad,None
```

```python
#can also use limit with method in fillna
new_df=df.fillna(method='ffill',limit=1)
new_df
#with limit clause we can specify the limitation of carrying forward
```

```python
#for a better guess of temp
#we can use interpolate method
new_df=df.interpolate()
new_df
#it basically takes mean of one prev and one forward
```

```python
#temperature on 4th Jan still not seems to be more precise
#as we dont have data of 2 and 3rd Jan
#to overcome it, we can use method=time
new_df=df.interpolate(method='time')
new_df
```

```python
#dropping null values in df
new_df=df.dropna()
new_df
#null records are dropped

#now my req is to drop a record that has all nulls
new_df=df.dropna(how='all')
new_df
#using all we dropped 9th jan data
```
